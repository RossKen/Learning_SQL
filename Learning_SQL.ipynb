```python
# what was the duration of the longest trip?

query = '''
SELECT id, duration 
FROM trips 
ORDER BY -duration 
LIMIT 1;
'''
# Note 'ORDER BY duration DESC' would have worked too

run_query(query)
```

```python
# top 10 longest journeys

query = '''
SELECT id, duration 
FROM trips 
ORDER BY -duration 
LIMIT 10;
'''
# Note 'ORDER BY duration DESC' would have worked too

run_query(query)
```

```python
# using WHERE statements

query = '''
SELECT *
FROM trips 
WHERE (duration > 9990) AND (sub_type = "Registered")
ORDER BY -duration ;
'''
# Note 'ORDER BY duration DESC' would have worked too

run_query(query)
```

```python
# How many trips taken by registered users?

query = '''
SELECT COUNT(*) AS "Total Trips by Registered Users"
FROM trips 
WHERE sub_type = "Registered";
'''

run_query(query)
```

```python
# Whta was the average trip duration

query = '''
SELECT AVG(duration) AS "Average Duration"
FROM trips;
'''

run_query(query)
```

```python
# Who has a longer avg duration - registered or casual riders?

query = '''
SELECT AVG(duration) AS "Average Duration"
FROM trips
WHERE sub_type = "Registered" ;
'''

run_query(query)

#Can deduce that casual riders are have higher average
```

```python
# Who has longer avg duration? - with GROUP BY

query = '''
SELECT sub_type, AVG(duration) AS 'Average Duration'
FROM trips
Group BY sub_type;
'''

run_query(query)
```

```python
# Which bike was used for the most trips?

query = '''
SELECT bike_number AS "Bike Number", COUNT(*) AS "Number of Trips"
FROM trips
GROUP BY bike_number
ORDER BY -"Number of Trips"
LIMIT 1;

'''

run_query(query)
```

```python
# Which station is the most frequent starting point?

query = '''
SELECT station AS "Station" , COUNT(*) AS "Count"
FROM trips
INNER JOIN stations 
ON trips.start_station = stations.id
GROUP BY stations.id
ORDER BY -"Count"
LIMIT 5;
'''

run_query(query)
```

```python
# Which stations are used most frequently for round trips?

query = '''
SELECT station AS "Station", Count(*) AS "Count"
FROM trips
INNER JOIN stations
ON trips.start_station = stations.id
WHERE start_station = end_station
GROUP BY stations.station
ORDER BY -"COUNT"
LIMIT 5;
'''

run_query(query)
```
